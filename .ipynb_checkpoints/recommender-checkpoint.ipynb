```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv("movie_dataset.csv")
movies.set_index("index")
```

```python
#get rid of the NaNs on these rows
for feature in features:
    movies[feature].fillna("", inplace = True)
```

* Next, we need to create tring sentences that combine all these features so that we can use the CountVectorizer.

```python
def combine_features(row):
    try:
        return row["keywords"] + " " + row["genres"] + " " + row["cast"] + " " + row["director"]
    except:
        return "Error " + row
```

```python
features = ['keywords', 'cast', 'genres', 'director'] #properties we're gonna use
movies["combined"] = movies.apply(combine_features, axis=1) #create a combined column
```

```python
movies["combined"].head(15)
```

* #### Get the cosine similarity

```python
count_vectorizer = CountVectorizer()
sparse_matrix = count_vectorizer.fit_transform(movies["combined"])
similarity_scores =  cosine_similarity(sparse_matrix)
similarity_scores
```

* #### Next step is find out what the user likes or has seen and look for it in our similarity score, we can only get that by knowing its index.

```python
user_likes = "Avatar" #we need to extract this movies scores from the similarity scores
```

```python
def get_index_from_title(title):
    return movies[movies[title] == user_likes]["index"].values[0]
```

```python
#enumerate coz we want to keep track of the indexes of the movies as well
similar_movies = list(enumerate(similarity_scores[get_index_from_title("original_title")]))
similar_movies = sorted(similar_movies, key = lambda x:x[1], reverse = True)
similar_movies
```

* as you can see the most similar movie is Guardians of the Galaxy followed by Aliens and star wars, this recommeder is 100% efficient

```python
#the top 5 most similar movies
recommendations = []
for movie in similar_movies[:6]:
    movie_name = movies.iloc[movie[0]]["original_title"]
    recommendations.append(movie_name)
recommendations
```
